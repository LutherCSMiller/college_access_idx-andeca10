```python
pgrant_percents=ipeds_raw[['UnitID', 'Institution Name', 'STABBR (HD2013)','PGRNT_P (SFA1314)']]
pgrant_percents.columns=[['UnitID','Institution_Name', 'State', 'Pell_Grant_Percents_1314']]
pgrant_percents=pgrant_percents.sort_values('Pell_Grant_Percents_1314', ascending=False)
pgrant_percents=pgrant_percents[(pgrant_percents.State == 'IA') | (pgrant_percents.State == 'MN') | (pgrant_percents.State == 'WI')]
```

```python
ax=pgrant_percents.plot(x='Institution_Name', y='Pell_Grant_Percents_1314', kind='barh', legend=False)
ax.set_xlabel('Percent Who Received Pell Grants 2013-2014')
```

## Plot in descending order the percentage of incoming first years who received Pell grants for colleges in MN, IA, and WI similar to and including Luther.

```python
Chart(pgrant_percents).mark_bar().encode(
    x='Pell_Grant_Percents_1314:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```

```python
grad_rate_6yr=ipeds_raw[['UnitID', 'Institution Name', 'STABBR (HD2013)','PGRNT_P (SFA1314)', 'GBA6RTT (DRVGR2014)']]
grad_rate_6yr.columns=[['UnitID', 'Institution_Name','State', 'Pell_Grant_%', '6yr_Grad_Rate']]
grad_rate_6yr.head()
```

```python
adj_pell_score = grad_rate_6yr[['UnitID','Institution_Name', 'State']]
adj_pell_score['Adj_Pell_Score'] = grad_rate_6yr['Pell_Grant_%']*(grad_rate_6yr['6yr_Grad_Rate']/100)
```

```python
adj_pell_score = adj_pell_score[(adj_pell_score.State == 'IA') | (adj_pell_score.State == 'MN') | (adj_pell_score.State == 'WI')].sort_values('Adj_Pell_Score', ascending=False)
adj_pell_score
```

## Plot, in descending order, the adjusted Pell Score based on the 6 year graduation rate of the colleges.

```python
Chart(adj_pell_score).mark_bar().encode(
    x='Adj_Pell_Score:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```

```python
net_prices=ipeds_raw[['UnitID', 'Institution Name', 'STABBR (HD2013)','NPT422 (SFA1314)',
       'NPT432 (SFA1314)']]
net_prices.columns=[['UnitID','Institution_Name', 'State', 'Ave_Net_Price_30k-48k','Ave_Net_Price_48k-75k']]
net_prices = net_prices[(net_prices.State == 'IA') | (net_prices.State == 'MN') | (net_prices.State == 'WI')]
```

```python
net_prices['Net_Price'] = net_prices[['Ave_Net_Price_30k-48k','Ave_Net_Price_48k-75k']].mean(axis=1)
net_prices=net_prices.sort_values('Net_Price', ascending=False)
net_prices
```

## Plot, in descending order, the average net price paid by those Title 4 aid recipients (Pell Grants) in the income tiers of 30,001-48,000 and 48,001-75000 at these comparable-to-and including Luther Midwestern colleges.

```python
Chart(net_prices).mark_bar().encode(
    x='Net_Price:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```

## Find the Average for each index, compare each institution to that.

```python
ave_score = adj_pell_score['Adj_Pell_Score'].mean()
ave_score
```

```python
adj_pell_score['Norm_Pell_Score'] = (adj_pell_score['Adj_Pell_Score'] / ave_score).round(2)
adj_pell_score
```

```python
ave_net_price = net_prices['Net_Price'].mean()
ave_net_price
```

```python
net_prices['Norm_Net_Price'] = (net_prices['Net_Price'] / ave_net_price).round(2)
net_prices
```

```python
ipeds_index = pd.merge(adj_pell_score, net_prices, left_on='UnitID', right_on='UnitID')
ipeds_index
```

```python
ipeds_index['Net_Index'] = ipeds_index[['Norm_Pell_Score', 'Norm_Net_Price']].mean(axis=1)
ipeds_index = ipeds_index.sort_values('Net_Index', ascending=False)
ipeds_index = ipeds_index[['Institution_Name_x', 'Net_Index']]
ipeds_index.columns = [['Institution_Name', 'Net_Index']]
```

## Plot the normalized average of both indices to create the total index for each college in the list.

```python
Chart(ipeds_index).mark_bar().encode(
    x='Net_Index:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```
