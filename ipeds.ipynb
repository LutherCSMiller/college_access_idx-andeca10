```python
pgrant_percents=ipeds_raw[['UnitID', 'Institution Name', 'STABBR (HD2013)','PGRNT_P (SFA1314)']]
pgrant_percents.columns=[['UnitID','Institution_Name', 'State', 'Pell_Grant_Percents_1314']]
pgrant_percents=pgrant_percents.sort_values('Pell_Grant_Percents_1314', ascending=False)
pgrant_percents=pgrant_percents[(pgrant_percents.State == 'IA') | (pgrant_percents.State == 'MN') | (pgrant_percents.State == 'WI')]
```

```python
ax=pgrant_percents.plot(x='Institution_Name', y='Pell_Grant_Percents_1314', kind='barh', legend=False)
ax.set_xlabel('Percent Who Received Pell Grants 2013-2014')
```

```python
Chart(pgrant_percents).mark_bar().encode(
    x='Pell_Grant_Percents_1314:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```

```python
grad_rate_6yr=ipeds_raw[['UnitID', 'Institution Name', 'STABBR (HD2013)','PGRNT_P (SFA1314)', 'GBA6RTT (DRVGR2014)']]
grad_rate_6yr.columns=[['UnitID', 'Institution_Name','State', 'Pell_Grant_%', '6yr_Grad_Rate']]
grad_rate_6yr.head()
```

```python
adj_pell_score = grad_rate_6yr[['UnitID','Institution_Name', 'State']]
adj_pell_score['Adj_Pell_Score'] = grad_rate_6yr['Pell_Grant_%']*(grad_rate_6yr['6yr_Grad_Rate']/100)
```

```python
adj_pell_score = adj_pell_score[(adj_pell_score.State == 'IA') | (adj_pell_score.State == 'MN') | (adj_pell_score.State == 'WI')].sort_values('Adj_Pell_Score', ascending=False)
adj_pell_score
```

```python
Chart(adj_pell_score).mark_bar().encode(
    x='Adj_Pell_Score:Q',
    y=Y('Institution_Name:O', sort=SortField(field='Institution_Name:O', order='descending'))
)
```
